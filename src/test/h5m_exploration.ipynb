```python
# edges is a list of 4 pairs of ints, this reorders the list so that [b, c] follows [a, b]
def reorder_edge_pairs(edges):
    print (edges)
    reordered = [edges[0]]
    for i in [0, 1, 2]:
        prev_edge = reordered[i]
        prev_end = prev_edge[1]
        for e in edges:
            if (e[0] == prev_end):
                reordered.append(e)
    return reordered
```

```python
def normal_from_two_edges(ordered_edges_list):
    edge_vectors = []
    for edge in ordered_edges_list[0:2]:
        coords = [np.array(ents.ents[i].coords) for i in edge]
        edge_vectors.append(coords[1] - coords[0])
        #print ("%s %s %s" % (edge, coords, edge_vectors[-1]))
    return np.cross(edge_vectors[0], edge_vectors[1])

normal_from_two_edges([[12, 3], [3, 1], [1, 6], [6, 12]])
```
